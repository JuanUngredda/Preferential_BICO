```python
import seaborn as sn
import numpy as np
import os
import pandas as pd
from torch import Tensor
import torch
import matplotlib.pyplot as plt
```

```python
from matplotlib.pyplot import figure
results = []

iters = range(10)
for s in iters:
    path = "data_for_plots/iter_{}.pkl".format(s)
    if os.path.exists(path):
        figure(figsize=(6, 6), dpi=80)
        df = pd.read_pickle(path)
        filter_mask = (df["parameters"][:,0] < 0.95) & (df["parameters"][:,0] > 0.04)
        # posterior samples
        plt.hist(df["parameters"][:,0][filter_mask], bins=60, density=True, label="$\Theta_{1}$", alpha=0.5)
        plt.hist(df["parameters"][:,1][filter_mask], bins=60, density=True, label="$\Theta_{2}$", alpha=0.5)
        
        plt.xlim(0,1)
        plt.legend(fontsize=23)
        
        plt.xticks(fontsize=20)
        plt.yticks(fontsize=20)
        plt.xlabel("$\Theta$", size=23)
        
        plt.savefig("/home/jungredda/Documents/GITHUB_REPOS/Preferential_BICO/plots/histogram_samples_{}.pdf".format(s), bbox_inches='tight')
        plt.show()
        
        figure(figsize=(6, 6), dpi=80)
        # objective space
        offset = 1
        plt.scatter(df["y"][:, 0] + offset, df["y"][:, 1] + offset, edgecolor="black", c="w", zorder=10, s=150)
        
        scaled_colors = torch.exp(df["acqvals"])
        if torch.sum(df["acqvals"]) != 0:
            color = (scaled_colors - torch.min(scaled_colors))/ (torch.max(scaled_colors) - torch.min(scaled_colors))

        else:
            color = df["acqvals"]
        
        print(df["acqvals"])
#         print(color)
#         raise
#         color_palette = plt.cm.plasma(np.linspace(0,1, len(color.numpy())))
        color_sorted = np.argsort(color)

        for i, idx in enumerate(color_sorted):
            edge = df["pairs"][idx]
            
            if color.numpy()[idx] ==0:
                color_value = plt.cm.viridis(np.linspace(0.,1.0, len(color.numpy()))[0])
                plt.plot([edge[0][0]+ offset, edge[1][0]+ offset], [edge[0][1]+ offset, edge[1][1]+ offset], c = color_value, alpha=0.5)
        
            else:
#                 color_value = plt.cm.viridis(np.linspace(0.,1.0, len(color.numpy()))[i])
                color_value = plt.cm.viridis(color.numpy()[idx])
                plt.plot([edge[0][0]+ offset, edge[1][0]+ offset], [edge[0][1]+ offset, edge[1][1]+ offset], c = color_value, alpha=0.5)
        
        plt.plot([edge[0][0]+ offset, edge[1][0]+ offset], [edge[0][1]+ offset, edge[1][1]+ offset], c = color_value)
        plt.scatter([df["best_pair"][0][0]+ offset, df["best_pair"][1][0]+ offset], [df["best_pair"][0][1]+ offset, df["best_pair"][1][1]+ offset], color="w", edgecolor="red", s=150,zorder=30, label="Best pair")
        
        plt.xlabel("$f_{1}$", size=23)
        plt.ylabel("$f_{2}$", size=23)
        plt.xticks(fontsize=20)
        plt.yticks(fontsize=20)
        
        plt.legend(fontsize=23, markerscale=2.)
        plt.savefig("/home/jungredda/Documents/GITHUB_REPOS/Preferential_BICO/plots/VoI_dm_{}.pdf".format(s), bbox_inches='tight')
        plt.show()
        print(color)
#         raise
```
